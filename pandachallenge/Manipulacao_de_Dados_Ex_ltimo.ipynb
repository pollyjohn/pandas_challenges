#### Tarefa 1. Importe o dataset e salve os dados em um dataframe

Os dados estão salvos no arquivo ***datasets/users_dataset.txt***.\
Salve os dados em uma variável de nome *users*.

````python
import pandas as pd
````

````python
users = pd.read_csv("users_dataset.csv", sep="|")
````

#### Tarefa 2. Descubra qual a idade média por ocupação.

*Dica: use a função groupby do Pandas na coluna occupation*

````python
# Seu código
users.groupby('occupation').agg({'age':'mean'})
````

#### Tarefa 3. Para cada ocupação descubra a idade mínima e máxima.

````python
# Seu código
(users.groupby("occupation").agg({"age": ["min", "max"]}))
````

#### Tarefa 4. Para cada combinação de ocupação e sexo, calcule a idade média.

Exemplo:

```
occupation     gender
administrator  F         40.638889
               M         37.162791
artist         F         30.307692
               M         32.333333
```

````python
# Seu código
(users.groupby(["occupation", "gender"]).agg({"age": "mean"}))
````

#### Tarefa 5. Para cada ocupação calcule a porcentagem de homens e mulheres.

Exemplo:

```
occupation     gender
administrator  F          45.569620
               M          54.430380
artist         F          46.428571
               M          53.571429
doctor         M         100.000000
educator       F          27.368421
               M          72.631579
```

````python

result = users.groupby(["occupation", "gender"]).agg({"user_id" : "count"})
result.groupby(level=0).apply(lambda x: 100 * x / x.sum())# Seu código
````

````python
(users.groupby(['occupation', 'gender'])['gender']
 .size().groupby(level=0)
 .apply(lambda x: 100 * x / x.sum())
 .round(2))
````

````python
occupation_gender = (users.groupby(["occupation", "gender"]).agg({"user_id": "count"})) 
````

````python
occupation_total_users = users.groupby("occupation").agg({"user_id":"count"})
````

````python
occupation_gender / occupation_total_users
````
